```python
import torch

# create two tensors of different sizes
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6, 7, 8])

# concatenate the tensors along a new dimension
c = torch.cat([a.unsqueeze(0), b.unsqueeze(0)], dim=0)


```

```python
# generate a random permutation of the concatenated tensor
permutation = torch.randperm(c.numel())

# select the random values from the concatenated tensor
```

```python
random_values = c.view(-1)[permutation[:2]]

# split the tensor back into its original parts
```

```python
random_value_a, random_value_b = torch.split(random_values, [1, 1])

print(random_value_a)
print(random_value_b)
```

```python
import torch

# create two tensors of different sizes
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6, 7, 8])

# pick a random value from each tensor
random_value_a = a[torch.floor(torch.rand(1) * len(a)).long()]
random_value_b = b[torch.floor(torch.rand(1) * len(b)).long()]

print(random_value_a)
print(random_value_b)
```
